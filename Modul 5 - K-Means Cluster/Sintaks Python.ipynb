```python
import pandas as pd
import numpy as np
from sklearn.cluster import KMeans
from matplotlib import pyplot as plt
from sklearn.preprocessing import StandardScaler
```

```python
Zulkarnaini = pd.read_excel("Data - Nilai machine learning.xlsx")
print(Zulkarnaini)
```

```python
Data_Zulkarnaini = Zulkarnaini[['UAS', 'UTS', 'Tugas 1', "IP"]]
wcss = []
for k in range(1,11):
    kmeans = KMeans(n_clusters=k, init='k-means++')
    kmeans.fit(Data_Zulkarnaini)
    wcss.append(kmeans.inertia_)
plt.figure(figsize=(10,6))
plt.plot(range(1,11), wcss, linewidth=2, marker= '8')
plt.xticks(np.arange(1,11,1))
plt.xlabel('K Values')
plt.ylabel('WCSS')
plt.show()
```

```python
mw=KMeans(n_clusters=3)
mw.fit(Data_Zulkarnaini)
y=mw.predict(Data_Zulkarnaini)
Zulkarnaini['label'] = y
print(Zulkarnaini.head(3))
```
