```python
# Impor library yang diperlukan
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsClassifier
from sklearn.metrics import accuracy_score, confusion_matrix

import matplotlib.pyplot as plt
import numpy
from sklearn import metrics
```

```python
Zulkarnaini = pd.read_csv("Data - diabetes.csv", sep = ';')
print(Zulkarnaini)
```

```python
#ZULKARNAINI
print(Zulkarnaini.describe())
```

```python
Zulkarnaini.info()
```

```python
target_name ='Outcome'
y = Zulkarnaini[target_name]
x = Zulkarnaini.drop(target_name, axis=1)
```

```python
#Split Data
#ZULKARNAINI

X_train, X_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state=42)
```

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
X_train = sc.fit_transform(X_train)
X_test = sc.transform(X_test)
print(X_train)
print(X_test)
```

# KNN

```python
k_list = list(range(3, 20, 2)) # k yang ganjil saja
k_list
```

```python
from sklearn.metrics import accuracy_score
for k in k_list:
    knn = KNeighborsClassifier(n_neighbors=k).fit(X_train, y_train)
    y_pred = knn.predict(X_test)
    accuracy = accuracy_score(y_test, y_pred)
    print(f"K = {k}, Akurasi = {accuracy: .4f}")
```

# Logistic Regression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt

```

```python
LR_Zulkarnaini = LogisticRegression(max_iter=1000, random_state=42)
LR_Zulkarnaini.fit(X_train, y_train)
```

```python
y_pred_lr = LR_Zulkarnaini.predict(X_test)

akurasi_lr = accuracy_score(y_test, y_pred_lr)
print("Tingkat Akurasi :%d persen" % (akurasi_lr*100))

print(classification_report(y_test, y_pred_lr))
```

```python
confusion_matrix_lr = metrics.confusion_matrix(y_test, y_pred_lr)
cm_display_lr = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_lr, display_labels=[False, True])
cm_display_lr.plot(cmap='Pastel1')
plt.show()
```

# Random Forest

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, classification_report, confusion_matrix
from sklearn import metrics
import matplotlib.pyplot as plt
```

```python
RF_Zulkarnaini = RandomForestClassifier(n_estimators=100, random_state=42)
RF_Zulkarnaini.fit(X_train, y_train)
```

```python
y_pred_rf = RF_Zulkarnaini.predict(X_test)

akurasi_rf = accuracy_score(y_test, y_pred_rf)
print("Tingkat Akurasi :%d persen" % (akurasi_rf*100))

print(classification_report(y_test, y_pred_rf))
```

```python
confusion_matrix_rf = metrics.confusion_matrix(y_test, y_pred_rf)
cm_display_rf = metrics.ConfusionMatrixDisplay(confusion_matrix=confusion_matrix_rf, display_labels=[False, True])
cm_display_rf.plot(cmap='Pastel1')
plt.show()
```
